```python
import seaborn as sns
import numpy as np
```

```python
# Import Datasets

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

## Categorical Plots

### Categorical Scatter Plot

- Stripplot
- Swarmplot

### Categorical Distribution Plots

- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

```python
sns.scatterplot(tips, x='total_bill', y="tip")
```

```python
# strip plot -> Axes level function

sns.stripplot(tips, x="day", y="total_bill")
```

```python
# catplot -> Figure level function

sns.catplot(tips, x="day", y="total_bill", kind='strip')
```

```python
# jitter

sns.catplot(tips, x="day", y="total_bill", kind='strip', jitter=0.2, hue='sex')
```

```python
# swarmplot

sns.catplot(tips, x="day", y="total_bill", kind='swarm', hue='sex')
```

```python
sns.swarmplot(tips, x="day", y="total_bill", hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot

sns.boxplot(tips, x='day', y='total_bill')
```

```python
# Using catplot

sns.catplot(tips, x='day', y='total_bill', kind='box')
```

```python
# hue

sns.boxplot(tips, x='day', y='total_bill', hue='sex')
```

```python
# single boxplot -> numerical col

sns.boxplot(tips, y='total_bill')
```

### Violinplot = (Boxplot + KDEplot)

```python
# violinplot

sns.violinplot(tips,x='day', y='total_bill')
```

```python
sns.catplot(tips,x='day', y='total_bill', kind='violin')
```

```python
# hue

sns.catplot(tips,x='day', y='total_bill', kind='violin', hue='sex', split=True)
```

```python
# barplot

sns.barplot(tips, x='sex', y='total_bill', hue='smoker', ci=None)
```

```python
sns.catplot(tips,kind='bar', x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
sns.catplot(tips,kind='bar', x='sex', y='total_bill')
```

```python
# point plot

sns.pointplot(tips, x='sex', y='total_bill', hue='smoker', ci=None)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# countplot

sns.countplot(tips, x='sex', hue='day')
```

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# faceting using catplot

sns.catplot(tips, x='sex', y='total_bill', col='smoker', kind='box', row='time')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level -> hue parameter is not available in axis level

sns.regplot(tips, x='total_bill', y='tip')
```

```python
sns.lmplot(tips, x='total_bill', y='tip', hue='sex')
```

```python
# residplot

sns.residplot(tips, x='total_bill', y='tip')
```

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot

sns.catplot(tips, x="sex", y="total_bill", kind='violin', col='day',row='time')
```

```python
g = sns.FacetGrid(tips, col='day', row='time', hue='smoker')
g.map(sns.scatterplot, 'tip', 'total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(iris, hue='species')
```

```python
# pair grid

g = sns.PairGrid(iris, hue='species')

# g.map

g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag

g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower

g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# vars

g = sns.PairGrid(iris, hue='species', vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

### JointGrid Vs Jointplot

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='hist', hue='sex')
```

```python
g = sns.JointGrid(tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.boxplot)
```
