### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> Axes Level -> Rectangle Shape -> Size and Hue

sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# relplot -> Figure Level -> Square Shape -> Size and Hue

sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
# Line Plot

gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Bangladesh']
```

```python
# Axes Level Function

sns.lineplot(temp_df, x='year', y='lifeExp')
```

```python
# Using relplot

sns.relplot(temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> Style

temp_df = gap[gap['country'].isin(['Bangladesh', 'Brazil', 'Germany'])]
temp_df
```

```python
sns.relplot(temp_df,kind='line', x='year', y='lifeExp', hue='country', style='continent' , size='continent')
```

```python
sns.lineplot(temp_df, x='year', y='lifeExp', hue='country', style='continent' , size='continent')
```

```python
# Facet plot -> Figure Level Function -> Work with relplot only.
# It will not work with scatterplot and lineplot

sns.relplot(tips, x='total_bill', y='tip', kind='scatter', col='sex', row='day')
```

```python
# col wrap

sns.relplot(gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

#### Figure Level -> displot
#### Axes level -> histplot -> kdeplot -> rugplot

```python
# Axes Level -> Plotting univariate histogram

sns.histplot(tips, x='total_bill')
```

```python
# Figure Level

sns.displot(tips, x='total_bill', kind='hist')
```

```python
# Bins Parameter

sns.displot(tips, x='total_bill', kind='hist', bins=50)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables
# Histogram acting like countplot
# countplot
sns.displot(tips, x='day', kind='hist')
```

```python
# Hue Parameter

sns.displot(tips, x='tip', kind='hist', hue='sex')
```

```python
# Element -> step

sns.displot(tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(titanic, x='age', kind='hist', element='step', hue='sex')
```

```python
# Faceting using col and row -> Will not work on histplot function.

sns.displot(tips, x='tip', kind='hist', element='step', col='sex')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.

sns.kdeplot(tips, x='total_bill')
```

```python
sns.displot(tips, x='total_bill', kind='kde')
```

```python
# Hue -> fill


sns.displot(tips, x='total_bill', kind='kde', hue='sex', fill=True, height=10, aspect=2)
```

```python
# Rugplot
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(tips, x='total_bill')
sns.rugplot(tips, x='total_bill')
```

```python
# Bivariate Histogram
# A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color.


sns.histplot(tips, x='total_bill', y='tip')
```

```python
sns.displot(tips, x='total_bill', y='tip', kind='hist')
```

```python
# Bivariate Kdeplot
# A bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap
# Plot rectangular data as a color-encoded matrix

temp_df = gap.pivot(index='country', columns='year', values='lifeExp')


# Axes Level Function (No Figure Level Function Available)
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot

temp_df = gap[gap['continent'] =='Europe'].pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(15,13))
sns.heatmap(temp_df, annot=True, linewidths=0.5, cmap='viridis')
```

```python
# Clustermap
# Plot a matrix dataset as a hierarchically-clustered heatmap.
# This function requires scipy to be available.

iris = px.data.iris()

plt.figure(figsize=(15,13))
sns.clustermap(iris.iloc[:,[0,1,2,3]], annot=True, linewidths=0.5, cmap='viridis')
```
