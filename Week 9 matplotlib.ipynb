### 2D Line plot

```python
# Plotting a Simple Function

price = [25000, 35000, 47000, 52000, 42000]
year = [2012, 2015, 2017, 2020, 2023]

plt.plot(year, price)
```

```python
# From a Pandas DataFrame


batsman = pd.read_csv("C:\\Users\\ACER\\Desktop\\DSMP1\\Datasets\\datasets-session-23\\sharma-kohli.csv")
batsman

plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# Plotting Multiple Plots

plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# Labels Title

plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# Colors (hex) and Line (width and style) and Marker (size)


plt.plot(batsman['index'], batsman['RG Sharma'], color='green')
plt.plot(batsman['index'], batsman['V Kohli'], color='black')

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'], color='red' , linestyle ='solid', linewidth = 3)
plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle ='dotted', linewidth = 3)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'], color='red', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)
plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# LLegend -> Location


plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit', color='red', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)
plt.plot(batsman['index'], batsman['V Kohli'], label='Kohli', color='blue', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
```

```python
# Limiting Axes

price = [20000, 30000, 25000, 50000, 35000, 5000000]
year = [2012, 2013, 2014, 2015, 2016, 2017]

plt.plot(year, price)
plt.ylim(0, 70000)
plt.xlim(2013, 2016)
```

```python
# Grid


plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit', color='red', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)
plt.plot(batsman['index'], batsman['V Kohli'], label='Kohli', color='blue', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
plt.grid()
```

```python
# Show

plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit', color='red', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)
plt.plot(batsman['index'], batsman['V Kohli'], label='Kohli', color='blue', linestyle ='solid', linewidth = 3, marker = 'o', markersize =10)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter Simple Function

x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0, 300, 50)
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter On Pandas Data

df = pd.read_csv("C:\\Users\\ACER\\Desktop\\DSMP1\\Datasets\\datasets-session-23\\batter.csv")
df = df.head(50)
```

```python
plt.scatter(df['avg'], df['strike_rate'], color = 'red', marker='+')


plt.title('Average and Strike Rate Analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
plt.figure(figsize=(15, 7.5))
plt.scatter(df['avg'], df['strike_rate'], color='blue', marker='o')

# Add player names as labels
for i in range(len(df)):
    plt.text(df['avg'][i], df['strike_rate'][i], df['batter'][i], fontsize=15)

plt.title('Average and Strike Rate Analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.tight_layout()
plt.show()
```

```python
# Size

tips = sns.load_dataset('tips')

plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20, color= 'red') # slower
```

```python
# ScatterPlot using plt.plot

plt.plot(tips['total_bill'], tips['tip'], 'o') # faster
```

```python
# plt.plot vs plt.scatter

"""
    plt.plot is faster to make but does not give customize option
    plt.scatter is slower to make but gives various customizable options
    
"""
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# Simple Bar Chart


children = [12, 16, 9, 19, 6]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color='black')
```

```python
# Bar Chart Using Data

```

```python
# Horizontal Bar Chart


plt.barh(colors, children, color='black')
```

```python
# Color and Label


df = pd.read_csv("C:\\Users\\ACER\\Desktop\\DSMP1\\Datasets\\datasets-session-23\\batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2 , label ='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, label ='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, label ='2017')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Performance by Year')
plt.legend()
plt.show()
```

```python
# A Problem

children = [12, 16, 9, 19, 6]
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']

plt.bar(colors, children)
plt.xticks(rotation=70)
plt.show
```

```python
# Stacked Bar Chart

plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2017'], bottom=df['2015'] + df['2016'], label='2017')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# Simple Data

data = [20,45,100,20,45]
subject = ['ENG', 'MAT', 'CHE', 'PHY', 'BIO']
plt.pie(data, labels= subject)
plt.show()
```

```python
# Dataset

df = pd.read_csv("C:\\Users\\ACER\\Desktop\\DSMP1\\Datasets\\datasets-session-23\\gayle-175.csv")
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# Percentage and Colors

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=['blue', 'green', 'yellow', 'pink', 'cyan', 'red'])
plt.show()
```

```python
# Explode Shadow

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.4,0,0,0,0,0.4], shadow=True)
plt.show()
```
